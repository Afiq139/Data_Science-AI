```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = [10,3]
plt.rcParams['figure.dpi'] = 200
```

```python
data = pd.read_csv('7_o_IMDB-Movie-Data.csv', nrows=50)
plt.hist(data['Rating'], bins = 20, edgecolor='white', color='orange')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Histogram Plot')
plt.show()
```
