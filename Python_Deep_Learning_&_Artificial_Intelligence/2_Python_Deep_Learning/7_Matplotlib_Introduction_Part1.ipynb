```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = [10,3]
plt.rcParams['figure.dpi'] = 100

data = pd.read_csv('7_o_IMDB-Movie-Data.csv', nrows=50)
data.head()
```

```python
print(data.columns)
plt.figure()
plt.plot(data['Rank'], data['Rating'])
plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Rank vs Rating')
plt.show()
```
